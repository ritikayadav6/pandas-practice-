```python
#area plot between two columns
#Data Visualization
import matplotlib.pyplot as plt
%matplotlib inline
y1 = cars['Engine Information.Engine Statistics.Horsepower']
y2 = cars['Engine Information.Engine Statistics.Torque']
x=range(5076)
plt.stackplot(x,y1,alpha=1,color='orange')
plt.stackplot(x,y2,alpha=0.5,color='green')
```

```python
#area plot between two columns
#Data Visualization
import matplotlib.pyplot as plt
%matplotlib inline
y1 = cars['Engine Information.Engine Statistics.Horsepower']
y2 = cars['Engine Information.Engine Statistics.Torque']
x=range(5076)
plt.plot(x,y1,linewidth=2.0,color='c')
plt.stackplot(x,y1,alpha=1,color='purple')
plt.plot(x,y2,linewidth=1.0,color='r')
plt.stackplot(x,y2,alpha=0.5,color='black')
```
